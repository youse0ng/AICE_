# 지역별(region) 인구의 히스토그램을 나타내고, 지역별 인구의 평균과 중앙값의 비교 결과를 고려하여 시사점을 설명하시오.


```python
dataframe = pd.read_csv(r'C:\Users\hyssk\AICE_\population.csv')
plt.figure(figsize=(20,6))
sns.barplot(dataframe,x='region',y='population')
plt.axhline(dataframe['population'].mean())
plt.ylabel("인구수")
plt.show()
```

# 지역별(region) 남녀비율의 히스토그램을 나타내고, 남녀비율의 평균과 중앙값의 비교 결과를 고려하여 시사점을 설명하시오.

```python
dataframe[['region','male','female']].mean()
dataframe['male'].mean()
plt.figure(figsize=(18,6))
plt.bar(x=dataframe['region'],height=dataframe['female'],label='female',color='red')
plt.bar(x=dataframe['region'],height=dataframe['male'],color='blue',label='male',width=0.5)
plt.legend()
plt.title('남/녀 지역별 성비 히스토그램')
plt.xlabel('지역')
plt.ylabel("인구수")
plt.axhline(dataframe['male'].mean(),color='blue',linestyle='--',linewidth=5)
plt.axhline(dataframe['female'].mean(),color='red',linestyle=':',linewidth=3)
plt.show()
```

# numpy의 np.random.normal()은 정규분포를 따르는 난수, np.random.exponential()은 지수분포를 따르는 난수를 생성해준다.

이 함수를 이용하여 평균 10, 표준편차 10인 정규 분포를 따르는 난수와, 평균 10인 지수분포를 따르는 난수를 각각 10,000개씩 발생시켜 이 값들을 상자그림(box plot)으로 비교하시오

```python
mean , std = 10 ,10 # 평균 10, 표준 편차 10
np.random.exponential
plt.figure()
plt.boxplot(np.random.normal(mean,std,10000))
plt.xlabel('정규분포(m:10,std:10)')
plt.show()
```

```python
# 정규 분포를 따르는 랜덤숫자를 발생
normal_data = np.random.normal(loc=10, scale=10, size=10000)

# 지수 분포를 따르는 랜덤숫자를 발생
exponential_data = np.random.exponential(scale=10, size=10000)

# 상자그림(box plot)으로 비교
plt.figure(figsize=(10, 6))
plt.boxplot([normal_data, exponential_data], labels=['Normal Distribution', 'Exponential Distribution'])
plt.title('Comparison of Normal and Exponential Distributions')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
